```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
data2 = df2.iloc[:,:12]
data2
```

```python
pca_data = preprocessing.scale(data2)
pca = PCA(n_components=12)
pc = pca.fit_transform(data2)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11','pc12']
pf = pd.DataFrame(data=pc,columns=names)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
loadings = pca.components_
num_pc = pca.n_components_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = data2.columns.values
loadings_df = loadings_df.set_index('variable')
```

```python
plt.rcParams['figure.figsize'] = (25,25)
ax = sb.heatmap(loadings_df, annot=True)
plt.show()
```

```python
from sklearn.cluster import KMeans

for j in range(1,11):
    inertia = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
        kmeans.fit(data2)
        inertia.append(kmeans.inertia_)


    plt.subplot(3,4,j)
    plt.plot(range(1, 11), inertia,marker = "8")
    plt.title('The Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')


plt.show()
```

```python
kmeans3 = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
kmeans3.fit(data2)
data2['cluster'] = kmeans3.labels_
print(kmeans3.cluster_centers_)
```
